### Explicit Solutions for \( \rho(t) \) and \( R(t) \) with Initial Condition at \( t = t_0 \)

We begin with the continuity equation for a perfect fluid:

\[
\frac{\dot{\rho}}{\rho} = \alpha H, \quad \alpha = -3(1 + w)
\]

Using the Hubble parameter definition \( H = \frac{\dot{R}}{R} \), and solving:

\[
\frac{d}{dt} \ln \rho = \alpha \frac{d}{dt} \ln R \Rightarrow \ln \rho = \alpha \ln R + \ln A \Rightarrow \rho = A R^\alpha
\]

From the Friedmann equation:

\[
H = \frac{\dot{R}}{R} = \sqrt{\frac{8\pi G}{3} \rho} = B \rho^{1/2}, \quad \text{with } B = \sqrt{\frac{8\pi G}{3}}
\]

Substitute \( \rho = A R^\alpha \) into Friedmann:

\[
\frac{\dot{R}}{R} = B (A R^\alpha)^{1/2} = B A^{1/2} R^{\alpha/2}
\Rightarrow \frac{dR}{dt} = B A^{1/2} R^{1 + \alpha/2}
\]

Solve the ODE:

\[
\int R^{-1 - \alpha/2} dR = B A^{1/2} \int dt
\]

This leads to the solution:

\[
R(t) = \left( \frac{3 H_0^2 (t - t_0)^2}{8 \pi G A} \right)^{\frac{1}{-3(1 + w)}}
\]

Using your boxed result:

\[
\rho(t) = \frac{1}{96\pi G (1 + w)^2} t^{-2}
\]

---

### Special Cases

- For \( w = 0 \) (matter-dominated):

\[
R(t) \propto t^{2/3}, \quad \rho(t) \propto t^{-2}
\]

- For \( w = \frac{1}{3} \) (radiation-dominated):

\[
R(t) \propto t^{1/2}, \quad \rho(t) \propto t^{-2}
\]

- For \( w = -\frac{1}{3} \) (curvature-like):

\[
R(t) \propto t, \quad \rho(t) \propto t^{-2}
\]

---

These match the standard cosmological results for a flat universe filled with a perfect fluid under general relativity.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python

class AnalyticalSolution:
    def __init__(self, w, H0=2.27e-18):
        self.w = w
        self.H0 = H0
        self.const = self.compute_constant()

    def compute_constant(self):
        denom = (36 * self.H0**2 * (1 + self.w)**2)**2
        base = 1 / denom
        exponent = 3 * (1 + self.w)
        return base**exponent

    def generate(self, t_min=1, t_max=10**18, num_points=1000):
        t_vals = np.linspace(t_min, t_max, num_points)
        R_vals = self.const * t_vals**(-6 * (1 + self.w))
        return t_vals, R_vals

    def plot(self):
        t, R = self.generate()
        plt.semilogy(t, R, label=f"w = {self.w}")
        plt.xlabel("Time t")
        plt.ylabel("Scale Factor R(t)")
        plt.title("Analytical Scale Factor Solution")
        plt.grid(True)
        plt.legend()
        plt.show()
```

```python


solution_1 = AnalyticalSolution(w=0) 

t1, R1 = solution_1.generate()

solution_1.plot()

```

```python

solution_2 = AnalyticalSolution(w=1) 

t2, R2 = solution_2.generate()

solution_2.plot()
```

```python
solution_3 = AnalyticalSolution(w=-0.33) 

t3, R3 = solution_3.generate()

solution_3.plot()
```

```python
params = {
            'w': 0,
            'G': 6.67430e-11,
            'H0': 2.27e-18,
            'R0': 1.0,
            't_span': (1e12, 1e18),
            'num_points': 1000,
            'atol': 1e-10,
            'rtol': 1e-8,
            'method': 'RK45'
        }
```

```python

class NumericalSolution:
    def __init__(self, params):
        
        self.params = params
        for iparam, param in params.items():
            setattr(self, iparam, param)
        
        self.A = np.sqrt(8 * np.pi * self.G / 3)
        self.rho0 = 3 * self.H0**2 / (8 * np.pi * self.G)
        self.y0 = [self.R0, self.rho0]

        # Time evaluation points
        self.t_eval = np.logspace(np.log10(self.t_span[0]), np.log10(self.t_span[1]), self.num_points)

    def set_param(self, parname, parval):
        setattr(self, parname, parval)

    def __call__(self, t, y):
        R, rho = y
        dR_dt = R * self.A * np.sqrt(rho)
        drho_dt = -3 * (1 + self.w) * self.A * rho**1.5
        return np.array([dR_dt, drho_dt])

    def solve(self):
        res = solve_ivp(fun=self,
                        t_span=self.t_span,
                        y0=self.y0,
                        vectorized=True,
                        method=self.method,
                        t_eval=self.t_eval,
                        atol=self.atol,
                        rtol=self.rtol)
        self.solution = res
        return res

    def plot(self):
        if not hasattr(self, 'solution'):
            self.solve()
        t = self.solution.t
        R = self.solution.y[0]
        rho = self.solution.y[1]

        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(t, R)
        plt.xscale("log")
        plt.xlabel("Time (s)")
        plt.ylabel("Scale Factor R(t)")
        plt.title("Scale Factor Evolution")

        plt.subplot(1, 2, 2)
        plt.plot(t, rho)
        plt.xscale("log")
        plt.yscale("log")
        plt.xlabel("Time (s)")
        plt.ylabel("Energy Density ρ(t)")
        plt.title("Energy Density Evolution")

        plt.tight_layout()
        plt.show()
```

```python
params_cosmo = {
    'w': 0,                           # Matter-dominated
    'G': 6.67430e-11,
    'H0': 2.27e-18,
    'R0': 1.0,
    't_span': (1e12, 1e18),
    'num_points': 1000,
    'atol': 1e-10,
    'rtol': 1e-8,
    'method': 'RK45'
}

cosmo = NumericalSolution(params_cosmo)

# Solve
result = cosmo.solve()

# Plot
cosmo.plot()

# Access results
t = result.t
R = result.y[0]
rho = result.y[1]
```
