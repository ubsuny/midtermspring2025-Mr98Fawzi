```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

# (A) Analytical Solver

## Mathmatical Derivation (Decouple the ODEs/ Direct Integration)

We study the evolution of the scale factor $ R(t) $ and energy density $\rho(t)$  for a flat FRW universe dominated by a fluid with equation of state $ p = w\rho $, using natural units $ G = 1 $.

---

#### Relating $ \rho $ and $ R $

We start from:

$$
\frac{d}{dt} \left\{ \ln \rho \right\} = \beta \frac{d}{dt} \left\{ \ln R \right\}, \quad \text{with} \quad \beta = -3(1 + w)
$$

$$
\Rightarrow \frac{d}{dt} \left\{ \ln \rho - \beta \ln R \right\} = 0 \quad \Rightarrow \quad \frac{\rho}{R^\beta} = C_2
$$

This implies:

$$
\rho = C_2 R^\beta \quad \Rightarrow \quad R = \tilde{C}_2 \rho^{1/\beta}
$$

---


From the Friedmann equation:

$$
\left( \frac{\dot{R}}{R} \right)^2 = \frac{8\pi}{3} \rho \quad \Rightarrow \quad \frac{\dot{R}}{R} = \sqrt{\frac{8\pi}{3}} \, \rho^{1/2} \equiv \alpha^{\frac{1}{2}}\rho^{1/2}
$$

The fluid equation:

$$
\dot{\rho} = \beta \frac{\dot{R}}{R} \rho = \beta \alpha^{\frac{1}{2}} \rho^{3/2}
$$

$$
\Rightarrow \frac{d\rho}{\rho^{3/2}} = \alpha^{\frac{1}{2}} \beta \, dt
$$

Integrate both sides:

$$
-2 \rho^{-1/2} = \alpha^{\frac{1}{2}} \beta t + \tilde{C}_1
$$

As $ \rho  \to \infty $, $ t \to 0 \Rightarrow \tilde{C}_1 = 0 $. Thus:

$$
\rho = \left( -\frac{1}{2} \alpha^{\frac{1}{2}} \beta t \right)^{-2} = \frac{1}{\left( \frac{1}{2} \alpha^{\frac{1}{2}} \beta t \right)^2}
$$

Substitute $ \alpha = \sqrt{\frac{8\pi }{3}} $, $ \beta = -3(1 + w) $:


$$
\rho(t) = \frac{1}{6\pi (1 + w)^2} \cdot t^{-2}
$$

---

#### Recovering the Scale Factor $ R(t) $

Using:

$$
R = \tilde{C}_2 \rho^{1/\beta}
$$

And:

$$
\rho(t) = \frac{1}{6\pi  (1 + w)^2} \cdot t^{-2}
$$

So:

$$
R(t) = \tilde{C}_2 \left( \frac{1}{6\pi (1 + w)^2} \right)^{1/\beta} \cdot t^{-2/\beta}
$$

$$
\Rightarrow R(t) \propto t^{2 / [3(1 + w)]}
$$

Normalize so that \( R(1) = 1 \), we set:

$$
R(t) = t^{\frac{2}{3(1 + w)}}, \quad \rho(t) = \frac{1}{6\pi  (1 + w)^2} \cdot t^{-2}
$$

---

## Our IVP :

$$
\boxed{
R(t) = t^{\frac{2}{3(1 + w)}}, \quad \rho(t) = \frac{1}{6\pi (1 + w)^2} \cdot t^{-2}
}
$$

$$
R(1) = 1, \quad \rho(1) = \frac{1}{6\pi (1 + w)^2}
$$


I designed a class `AnalyticalSolution` that takes the equation of state parameter $ w $ and computes the time evolution of the scale factor $ R(t) $ and energy density $ \rho(t) $ from the early universe $ t \approx 0 $ up to $ t = 10 $, where the present day is represented by $ t = 1 $.

I use NumPy broadcasting to evaluate $ R(t) $ and $ \rho(t) $ over a time array `t_vals` based on the analytical forms:


```python
class AnalyticalSolution:
    def __init__(self, w,t_min=1e-4,t_max=10,num_points=1000):
        self.w = w

        self.R0=1
        self.rho_0 = 1/(6*np.pi*(1 + self.w)**2)
        
        self.t_min=t_min
        self.t_max=t_max
        self.num_points=num_points
        
        self.t_vals = np.linspace(t_min, t_max, num_points)

    def generate(self):
        t_vals =self.t_vals
        R_vals = self.R0 * t_vals**(2 / (3 * (1 + self.w)))
        rho_vals = self.rho_0 * t_vals**(-2)
        return t_vals, R_vals, rho_vals
```

Here I define plotter functions for the scale factor $ R(t) $ and energy density $ \rho(t) $.

The energy density is plotted on a logarithmic scale due to its steep decay in time.


```python
def plot_R(t,R,w):
    plt.plot(t, R, label=f"w = {w}")
    plt.plot(1,1, 'ro',markersize=4)
    plt.xlabel("t")
    plt.ylabel("R(t)")
    plt.legend()
    plt.grid(True)

def plot_rho(t,rho,w):
    plt.semilogy(t, rho, label=f"w = {w}")
    plt.xlabel("t")
    plt.ylabel(r"$log(\rho(t))$")
    plt.legend()
    plt.grid(True)
```

```python
solution_1 = AnalyticalSolution(w=0) 
t1, R1, rho_1 = solution_1.generate()

solution_2 = AnalyticalSolution(w=1) 
t2, R2, rho_2 = solution_2.generate()

solution_3 = AnalyticalSolution(w=-0.33) 
t3, R3, rho_3 = solution_3.generate()

```

```python
plt.figure(dpi=500)
plot_R(t1,R1,w=0)
plot_R(t2,R2,w=1)
plot_R(t3,R3,w=-0.33)
plt.title("Multiple scale factor R(t) Curves (Analytical)")
plt.show()
```

```python
plt.figure(dpi=500)
plot_rho(t1,rho_1,w=0)
plot_rho(t2,rho_2,w=1)
plot_rho(t3,rho_3,w=-0.33)
plt.title(r"Multiple Density $\rho(t)$ Curves (Analytically)")
plt.show()
```

### NumericalSolution: Inheriting from AnalyticalSolution

We define a class `NumericalSolution` that inherits from `AnalyticalSolution` to numerically solve the coupled ODE system for the scale factor $R(t)$ and  density $\rho(t)$.


The inheritance allows `NumericalSolution` to reuse:
- The equation of state parameter $w$
- The time interval $[t_{\min}, t_{\max}]$
- The number of time points
- The analytical expression for $\rho_0$ (used as the initial condition)

We define:
- The initial time $t_0 = 1$ to represent "today"
- Initial conditions $R_0 = 1$, $\rho_0 = \rho(t=1) = \frac{1}{6\pi(1+w)^2}$
- The dynamical system:


This system is solved using `scipy.integrate.solve_ivp`, both:
- **Forward in time**: from $t_0 \to t_{\max}$
- **Backward in time**: from $t_0 \to t_{\min}$



# (B) Part I. Numerical Solver

```python
class NumericalSolution(AnalyticalSolution):
    def __init__(self, w, t_min=1e-4, t_max=10, num_points=1000,
                 atol=1e-10, rtol=1e-8, method='RK45'):
        super().__init__(w, t_min, t_max, num_points)

        self.atol = atol
        self.rtol = rtol
        self.method = method

        self.A = np.sqrt(8 * np.pi / 3)
        self.B = -3 * (1 + self.w)

        self.t0 = 1.0
        self.R0 = 1.0
        self.rho0 = self.rho_0
        self.y0 = [self.R0, self.rho0]

        self.t_vals_forward = np.linspace(self.t0, self.t_max, self.num_points)
        self.t_vals_backward = np.linspace(self.t0, self.t_min, self.num_points)  # normal order

    def __call__(self, t, y):
        R, rho = y
        dR_dt = self.A * R * np.sqrt(rho)
        H = dR_dt / R
        drho_dt = self.B * H * rho
        return np.array([dR_dt, drho_dt])

    def solve(self):
        sol_forward = solve_ivp(fun=self,
                                t_span=(self.t0, self.t_max),
                                y0=self.y0,
                                method=self.method,
                                t_eval=self.t_vals_forward,
                                atol=self.atol,
                                rtol=self.rtol)
    
        # Backward integration: t_span is decreasing, t_eval must also be decreasing
        t_eval_backward = np.linspace(self.t0, self.t_min, self.num_points)  # descending
        sol_backward = solve_ivp(fun=self,
                                 t_span=(self.t0, self.t_min),
                                 y0=self.y0,
                                 method=self.method,
                                 t_eval=t_eval_backward,
                                 atol=self.atol,
                                 rtol=self.rtol)
    
        # Reverse backward results to chronological order
        t_b = sol_backward.t[::-1]
        y_b = sol_backward.y[:, ::-1]
    
        # Combine (avoid duplicate at t0)
        t_vals = np.concatenate((t_b[:-1], sol_forward.t))
        y_vals = np.hstack((y_b[:, :-1], sol_forward.y))
    
        R_vals, rho_vals = y_vals
        return t_vals, R_vals, rho_vals
```

```python
num_sol_1 = NumericalSolution(w=0)
t_vals_1, R_vals_1, rho_vals_1 = num_sol_1.solve()

num_sol_2 = NumericalSolution(w=1)
t_vals_2, R_vals_2, rho_vals_2 = num_sol_2.solve()

num_sol_3 = NumericalSolution(w=-0.33)
t_vals_3, R_vals_3, rho_vals_3 = num_sol_3.solve()
```

```python
plt.figure(dpi=500)
plot_R(t_vals_1, R_vals_1, w=0)
plot_R(t_vals_2, R_vals_2, w=1)
plot_R(t_vals_3, R_vals_3, w=-0.33)
plt.title("Multiple Scale Factor R(t) Curves (Numerically)")
plt.show()
```

```python
plt.figure(dpi=500)
plot_rho(t_vals_1, rho_vals_1, w=0)
plot_rho(t_vals_2, rho_vals_2, w=1)
plot_rho(t_vals_3, rho_vals_3, w=-0.33)
plt.title(r"Multiple Density $\rho(t)$ Curves (Numerically)")
plt.show()
```

# (B) Part II. Compare Analytical Vs Numerical Sol

```python

colors = ["tab:blue", "tab:green", "tab:red"]
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10), dpi=500)
titles = [r"$w = 0$", r"$w = 1$", r"$w = -\frac{1}{3}$"]

# Scale Factor R(t)
axes[0, 0].plot(t1, R1, label="Analytical", linestyle="--", color=colors[0])
axes[0, 0].scatter(t_vals_1, R_vals_1, s=10, label="Numerical", color="black")
axes[0, 0].set_title(titles[0])
axes[0, 0].set_ylabel(r"Scale Factor $R(t)$")
axes[0, 0].grid(True)
axes[0, 0].legend()

axes[0, 1].plot(t2, R2, linestyle="--", color=colors[1])
axes[0, 1].scatter(t_vals_2, R_vals_2, s=10, color="black")
axes[0, 1].set_title(titles[1])
axes[0, 1].grid(True)

axes[0, 2].plot(t3, R3, linestyle="--", color=colors[2])
axes[0, 2].scatter(t_vals_3, R_vals_3, s=10, color="black")
axes[0, 2].set_title(titles[2])
axes[0, 2].grid(True)

# rho(t) with log scale
axes[1, 0].plot(t1, rho_1, linestyle="--", color=colors[0])
axes[1, 0].scatter(t_vals_1, rho_vals_1, s=10, color="black")
axes[1, 0].set_ylabel(r"Density $log(\rho(t))$")
axes[1, 0].set_xlabel("Time $t$")
axes[1, 0].set_yscale("log")
axes[1, 0].grid(True, which='both', axis='y')

axes[1, 1].plot(t2, rho_2, linestyle="--", color=colors[1])
axes[1, 1].scatter(t_vals_2, rho_vals_2, s=10, color="black")
axes[1, 1].set_xlabel("Time $t$")
axes[1, 1].set_yscale("log")
axes[1, 1].grid(True, which='both', axis='y')

axes[1, 2].plot(t3, rho_3, linestyle="--", color=colors[2])
axes[1, 2].scatter(t_vals_3, rho_vals_3, s=10, color="black")
axes[1, 2].set_xlabel("Time $t$")
axes[1, 2].set_yscale("log")
axes[1, 2].grid(True, which='both', axis='y')

plt.tight_layout()
plt.suptitle("Analytical (dashed color lines) vs Numerical (black scatter)", fontsize=16, y=1.02)
plt.show()
```
